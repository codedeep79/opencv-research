**Face Detection**

```python
faceCascade = cv2.CascadeClassifier('/content/data/haarcascade_frontalface_default.xml')
def detectFace(img):
  faceImg = img.copy()
  faceRects = faceCascade.detectMultiScale(faceImg,scaleFactor=1.2, minNeighbors=5) 
  for (x,y,w,h) in faceRects: 
    cv2.rectangle(faceImg, (x,y), (x+w,y+h), (255,255,255), 10) 
  return faceImg
result = detectFace(people1)
plt.imshow(result,cmap='gray')
```

```python
result = detectFace(people2)
plt.imshow(result,cmap='gray')
```

```python
result = detectFace(people3)
plt.imshow(result,cmap='gray')
```
